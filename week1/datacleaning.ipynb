```python
import numpy as np
import pandas as pd


# load dataset
df= pd.read_csv('penguins.csv')

print(" \n head \n")
# Display the first 5 rows
print(df.head())

print(" \n info \n")
# Display summary information
print(df.info())

print(" \n describe\n")
# Display statistical summary of numerical columns
print(df.describe())

df.sample(5) #selects random selected number of rows

#isnull finds missing values and sum adds all the values for that column
missing_values = df.isnull().sum()
print(missing_values)

df.fillna(value=0, inplace=True) #fills missing values
print(missing_values)

df.drop_duplicates(inplace=True) # Remove duplicate rows 

df.rename(columns={'species': 'species_name'}, inplace=True) 
#Rename a column {'OldName': 'NewName'}
df.head(3) #testing the column name is changed or not

print(df.info()) 
```
