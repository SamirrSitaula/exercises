```python
-Core part of data preprocessing or ML PIPELINE

1. GENERATE SYNTHETIC DATA
2. EXTRACT/ ENCODING
    A) STATISTICAL FEATURES[MEAN, STD...]
    B) DATA/ TIME FEATURES
    C) ENCODING CATEGORICAL FEATURES
    D) TEXT FEATURE EXTRACTION USING TF-IDF
```

```python
#importing basic library
import numpy as np #for all kind of numeric operation
import pandas as pd # handling tabular data

np.random.seed(42) #generating random data
```

```python
data = { 
    'id': range(1,101),
    'age': np.random.randint(18,60,100), #random intiger
    'income': np.random.normal(50000, 15000, 100),
    'signup_date': pd.date_range(start = '2002-01-01', periods = 100, freq='D'),
    'category': np.random.choice(['A', 'B', 'C'],100),
    'feedback': np.random.choice(['Great Product',' Not Good', 'Average Experience','Bad Service'],100)
}
#data

    
```

```python
#arranging the data
df = pd.DataFrame(data)
df
```

```python
#statistical Feature Extraction
df['income_per_age'] = df['income']/df['age']
```

```python
#z-score of Income, how far eacg income value is from mean
df['income_zscore'] = (df['income']-df['income'].mean())/df['income'].std()
df
```

```python
df.head()
```

```python
df.describe()
```

```python
#describe things for selected column
df[['age', 'income','income_per_age','income_zscore']].describe()
```

```python
#Date/ Time Feature Exxtraction
df['signup_day'] = df['signup_date'].dt.day #day extraction

df['signup_month'] = df['signup_date'].dt.month #month extraction
df
```

```python
df.head()
```

```python
df['signup_weekday'] = df['signup_date'].dt.weekday #weekday extraction
df
```

```python
df['days_from_signup'] = (pd.Timestamp.today() - df['signup_date']).dt.days/365 #weekday extraction
df
```

```python
#Encode Categorical Variables
df_encoded=pd.get_dummies(df, columns = ['category'], prefix='cat')
```

```python
df_encoded.head()
```

```python
#text feature extraction(TF-IDF)
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['feedback'])
```

```python
print(tfidf_matrix)
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(),columns=vectorizer.get_feature_names_out())
```

```python
df_combined = pd.concat([df,tfidf_df], axis = 1)
```

```python
print(df_combined[['feedback']+ list(tfidf_df.columns)].head()) #BOG --> Bag of words
```
