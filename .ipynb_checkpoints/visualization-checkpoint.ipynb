```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample DataFrame
np.random.seed(42) #random value wont change at every execution, remains same as first execution
data = pd.DataFrame({
    'Age': np.random.randint(20, 60, 100),
    'Salary': np.random.randint(30000, 100000, 100)
})

plt.figure(figsize=(15, 5))  # Create figure with specified size

# 1 **Scatter Plot**
plt.subplot(1, 3, 1)  # (rows, columns, position)
sns.scatterplot(x=data['Age'], y=data['Salary'], color='red')
plt.title('Age vs Salary (Scatter Plot)')
plt.xlabel('Age')
plt.ylabel('Salary')

# 2 **Histogram**
plt.subplot(1, 3, 2)
sns.histplot(data['Age'], bins=6, kde=True, color='green')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# 3 **Box Plot**
plt.subplot(1, 3, 3)
sns.boxplot(y=data['Salary'], color='orange')  # Fixed incorrect syntax
plt.title('Box Plot of Salary')
plt.xlabel('Salary')

#bar on fourth
plt.subplot(2, 3, 4)  
sns.barplot(x='Age', y='Salary', data=data, color='blue')
plt.title('Salary vs Age (Bar Plot)')
plt.xlabel('Age')
plt.ylabel('Salary')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```
